```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
import pandas as pd
X=[]
Y=[]
T=[]

F,N=map(int,input().split())
for i in range(N):
    il=input().split()
    il=[float(f) for f in il]
    X.append(il[:,-1])
    Y.append(il[-1])
t=int(input())
for i in range(t):
    il=input().split()
    il=[float(f) for f in il]
    T.append(il)

X=np.array(X)
Y=np.array(Y)
T=np.array(T)
    
poly = PolynomialFeatures(degree=F)
XP = poly.fit_transform(X)
TP= poly.fit_transform(T)

clf = linear_model.LinearRegression()
clf.fit(XP, Y)
print clf.predict(TP)
```

```python
# Complete the counts function below.
import sys
def counts(nums, maxes):
    # l=[None]*len(maxes)
    
    count=0
    for i in maxes:
        for j in nums:
            if maxes[i]>nums[j]:
                count+=1
        print(count)   
        count=0

maxes=[]
nums=[]
data= sys.stdin.readlines()
N= int(data[0])
# data.pop(0)
for i in range(N):
    nums.append( int(data[i+1]))
M =int(data[N+2])    
# data.pop(N+1)    
for j in range(M): 
    maxes.append(int(data[-j]))
counts(nums, maxes)    

print(maxes)
```
